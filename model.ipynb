```python
import pandas as pd
import numpy as np
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("st.csv",low_memory=False)
df=df.drop(["Unnamed: 10","Unnamed: 11","Unnamed: 12","Unnamed: 13","Unnamed: 14","Unnamed: 15","Unnamed: 16","Unnamed: 17"],axis="columns")
df=df.dropna()
df
```

```python
# df['Unix Timestamp'] = pd.to_datetime(df['Date'], format="%d-%m-%Y").astype(int) / 1e9
df
```

```python
df=df[~((df['Date']=="Date"))]
df
```

```python
df=df[~(df['Date']=="Zee Entertainment Enterprises Ltd.")]
```

```python
df['Date1'] = pd.to_datetime(df['Date'], format="%d-%m-%Y")
df
```

```python
df["Date"] = pd.to_datetime(df['Date1'], format='%D-%M-%Y')
df["Unix_Date"] = df['Date'].apply(lambda x: x.timestamp())/1e7
```

```python
df
```

```python
df=df.drop(["Date","Date1"],axis=True)
df
```

```python
dummies=pd.get_dummies(df["Symbol"]).astype(int)
df=pd.concat([df,dummies],axis="columns")
df=df.drop(["Symbol"],axis="columns")
df
```

```python
df=df.drop(["Last"],axis="columns")
```

```python
df["Volume"].max()
```

```python
df["Prev Close"]=pd.to_numeric(df["Prev Close"])/100
df["Open"]=pd.to_numeric(df["Open"])/100
df["High"]=pd.to_numeric(df["High"])
df["Low"]=pd.to_numeric(df["Low"])
df["Close"]=pd.to_numeric(df["Close"])/100
df["VWAP"]=pd.to_numeric(df["VWAP"])/100
df["Volume"]=pd.to_numeric(df["Volume"])/1e5
df
```

```python
y=df[["High","Low"]]
y
```

```python
X=df.drop(["High","Low"],axis="columns")
X
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=48)
```

```python
X_train.shape,y_train.shape
```

```python
model=keras.Sequential([
    keras.layers.Flatten(input_shape=(71,)),
    keras.layers.Dense(500,activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(100,activation='relu'),
    keras.layers.Dropout(0.2),
    keras.layers.Dense(2,activation='relu')
])
model.compile(loss="MSE",optimizer='adam',metrics=['accuracy'])
```

```python
model.fit(X_train,y_train,epochs=5)
```

```python
model.evaluate(X_test,y_test)
```

```python
model.predict(X_test)
```

```python
def predict_price(Company,prev_close,open,close,VWAP,Volume,date):    
    loc_index = np.where(X.columns==Company)[0][0]
    x = np.zeros(len(X.columns))
    x[0] = prev_close/100
    x[1] = open/100
    x[2] = close/100
    x[3]=VWAP/100
    x[4]=Volume/1e5
    x[5]=pd.to_datetime(date).timestamp()/1e7
    print(x[5])
    if loc_index >= 0:
        x[loc_index] = 1
    # x=x.reshape(1,71)
    return (model.predict([x])[0]).astype(int)
```
